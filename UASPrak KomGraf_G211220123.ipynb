```python
# Import File
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
```

```python
# Show File
dataset = pd.read_csv('datavito.csv')
dataset.head(15)
```

```python
# Distribusi Purchase Price
plt.figure(figsize=(10,7))
plt.hist(dataset['Purchase Price'], bins=15, edgecolor='white', color='red')
plt.title("History of Purchase Price",fontsize=20)
plt.xlabel("Purchase Price", fontsize=18)
plt.ylabel("Frequency", fontsize=18)

plt.show()
```

```python
# Distribusi Current Price
plt.figure(figsize=(10,7))
plt.hist(dataset['Current Price'], bins=15, edgecolor='white', color='yellow')
plt.title("History of Current Price",fontsize=20)
plt.xlabel("Current Price", fontsize=18)
plt.ylabel("Frequency", fontsize=18)

plt.show()
```

```python
# Distribusi Gain/Loss
plt.figure(figsize=(10,7))
plt.hist(dataset['Gain/Loss'], bins=15, edgecolor='white', color='green')
plt.title("History of Gain/Loss",fontsize=20)
plt.xlabel("Gain/Loss", fontsize=18)
plt.ylabel("Frequency", fontsize=18)

plt.show()
```

```python
# Top 5 Dividend/Share
dataset_sorted = dataset.sort_values(by='Dividend/share', ascending=False)
top_5 = dataset_sorted.head(5)[['Stock Name','Symbol',]]
print(top_5)
```

```python
# Top 5 Market Value
dataset_sorted = dataset.sort_values(by='Market Value', ascending=False)
top_5 = dataset_sorted.head(5)[['Stock Name','Symbol',]]
print(top_5)
```

```python
# Top 5 Annual Yield
dataset_sorted = dataset.sort_values(by='Annual Yield', ascending=False)
top_5 = dataset_sorted.head(5)[['Stock Name','Symbol',]]
print(top_5)
```
